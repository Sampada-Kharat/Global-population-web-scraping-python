```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
url="https://www.worldometers.info/world-population/population-by-country/"
res=requests.get(url)
res
```

```python
res.text
```

```python
res.content
```

```python
soup=BeautifulSoup(res.content)
```

```python
soup.prettify()
```

```python
soup.title.text
```

```python
title=soup.title
title
```

```python
Population_data=soup.find("table")
Population_data
```

```python
data=Population_data.find("tbody")
data
```

```python
rows=data.find_all("tr")
rows
```

```python
header_row = data.find("tr")
headers = [cell.text.strip() for cell in header_row.find_all(["th", "td"])]
```

```python
info = []
for i in rows:  # rows = soup.find_all("tr")
    cols = i.find_all("td")
    if not cols:  # skip rows without <td>
        continue
    row_data = [j.text.strip() for j in cols]  # clean text
    info.append(row_data)
```

```python
print(info)
```

```python
Series_No=[]
Country_or_Dependency=[]
Population=[]
Yealy_Change=[]
Net_Change=[]
Density_per_square_kilometer=[]
Land_Area_square_kilometer=[]
Migrants=[]
Fertility_Rate=[]
Median_Age=[]
Urban_Population_percentage=[]
World_Share=[]

Series_No = [i[0] for i in info]
Country_or_Dependency = [i[1] for i in info]
Population = [i[2] for i in info]
Yealy_Change = [i[3] for i in info]
Net_Change = [i[4] for i in info]
Density_per_square_kilometer = [i[5] for i in info]
Land_Area_square_kilometer = [i[6] for i in info]
Migrants = [i[7] for i in info]
Fertility_Rate = [i[8] for i in info]
Median_Age = [i[9] for i in info]
Urban_Population_percentage = [i[10] for i in info]
World_Share = [i[11] for i in info]

print(Series_No)
print(Country_or_Dependency)
print(Population)
print(Yealy_Change)
print(Net_Change)
print(Density_per_square_kilometer)
print(Land_Area_square_kilometer)
print(Migrants)
print(Fertility_Rate)
print(Median_Age)
print(Urban_Population_percentage)
print(World_Share)
```

```python
df = pd.DataFrame(info, columns=[
    "Series_No",
    "Country_or_Dependency",
    "Population",
    "Yealy_Change",
    "Net_Change",
    "Density_per_square_kilometer",
    "Land_Area_square_kilometer",
    "Migrants",
    "Fertility_Rate",
    "Median_Age",
    "Urban_Population_percentage",
    "World_Share"
])
```

```python
df.to_csv("population_data.csv", index=False)
```
